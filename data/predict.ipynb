```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')

# Load the dataset
df = pd.read_excel('/content/drive/My Drive/dataset.xlsx', sheet_name='Sheet1')

target_column = 'Commissioning of individual and multi-apartment residential buildings'

# Initialize a list to store forecasts
forecast_list = []

# Loop through each region and fit the ARIMA model
for region in df['Republic of Kazakhstan'].unique():
    region_data = df[df['Republic of Kazakhstan'] == region].groupby('Year')[target_column].mean()

    # Fit the ARIMA model
    model = ARIMA(region_data, order=(1, 1, 1))
    model_fit = model.fit()

    # Forecast for the next 3 years
    forecast = model_fit.forecast(steps=3)
    print(forecast)
    # Store the results in the list
    forecast_list.append({
        'Region': region,
        '2025': forecast[3],
        '2026': forecast[4],
        '2027': forecast[5]
    })

    # Plot the historical data and forecast for the current region
    plt.figure(figsize=(10, 5))
    plt.plot(region_data.index, region_data, label='Historical Data')
    plt.plot([2025, 2026, 2027], forecast, label='Forecast', marker='o')
    plt.title(f'ARIMA Forecast for {region}')
    plt.xlabel('Year')
    plt.ylabel('Commissioning of individual and multi-apartment residential buildings')
    plt.legend()
    plt.grid()
    plt.show()

# Convert the list to a DataFrame
forecast_results = pd.DataFrame(forecast_list)

# Print forecast results for all regions
print("Forecast results for 2025, 2026, 2027 by region:")
print(forecast_results)
```
